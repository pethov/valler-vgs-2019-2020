## nummerisk derivasjon


```python
from pylab import*

data=loadtxt('tekster.txt',skiprows=1, delimiter=',')

tid= data[:,0]
hoyde= data[:,2]
fart=[]
aks=[]

plot(tid,hoyde)
xlabel('minutter')
ylabel('meter')
show()


tiden=list(tid)
hoyden= list(hoyde)

for i in range (len(hoyden)-1):
    sder=(hoyden[i+1]-hoyden[i])/(tid[i+1]-tid[i])
    fart.append(sder)
    
tiden.pop(-1)
plot(tiden,fart)
xlabel('minutter')
ylabel('m/s')
show()

for i in range(len(fart)-1):
    vder=(fart[i+1]-fart[i])/(tid[i+1]-tid[i])
    aks.append(vder)
    
tiden.pop(-1)
plot(tiden,aks)
xlabel('minutter')
ylabel('m/s^2')
show()
```
