# regresjon

en statistitsk teknik for en varibiel utifra andre variabler. hvor man danner en tilnærmet funksjon mellom punktene.

```python
from pylab import*
from sklearn.metrics import r2_score

T = array([0,20,40,60,80,100])
sol_NH3 = array([88.5, 56.0, 34.0, 20.0, 11.0, 7.0])
sol_NaCl = array([35.7, 35.9, 36.4, 37.1, 38.0, 39.2])

grad = 2
reg_NaCl = polyfit(T, sol_NaCl, grad)
reg_NH3 = polyfit(T, sol_NH3, 3)

x= linspace(0,150,10000)
a= reg_NaCl[0]
b= reg_NaCl[1]
c= reg_NaCl[2]
d= reg_NH3[0]
e= reg_NH3[1]
f= reg_NH3[2]
g= reg_NH3[3]
y= a*x**2 + b*x + c
z_R2= d*T**3 + e*T**2 +f*T + g
z= d*x**3 + e*x**2 +f*x + g

R2= r2_score(sol_NH3, z_R2)

scatter(T, sol_NaCl)
scatter(T, sol_NH3)
plot(x,y, color='m')
plot(x,z, color='c')

print(R2)
```
